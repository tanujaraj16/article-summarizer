```python
from transformers import pipeline
import torch

# Use GPU if available
device = 0 if torch.cuda.is_available() else -1
print("Torch CUDA available:", torch.cuda.is_available())
```

```python
# Load GPT-2 model for text generation
generator = pipeline("text-generation", model="gpt2", device=device)
print("✅ GPT-2 is ready")
```

```python
# Example prompt
prompt = "The future of artificial intelligence in healthcare is"

# Generate text
result = generator(
    prompt,
    max_length=150,
    temperature=0.8,
    top_k=40,
    top_p=0.9,
    do_sample=True
)[0]['generated_text']

print("=== Generated Paragraph ===\n")
print(result)
```

```python
# Ask user for prompt
user_prompt = input("Enter your topic: ")

# Generate text
generated_text = generator(
    user_prompt,
    max_length=150,
    temperature=0.8,
    top_k=40,
    top_p=0.9,
    do_sample=True
)[0]['generated_text']

print("\n=== GPT-2 Generated Paragraph ===\n")
print(generated_text)
```

```python
# STEP 1: Install dependencies
!pip install torch torchvision pillow matplotlib --quiet

# STEP 2: Import libraries
import torch, torchvision
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# STEP 3: Helper functions
def load_image(path, size=512):
    image = Image.open(path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((size, size)),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ])
    return transform(image).unsqueeze(0).to(device)

def im_convert(tensor):
    image = tensor.clone().detach().cpu().squeeze(0).numpy().transpose(1,2,0)
    mean = np.array([0.485,0.456,0.406])
    std  = np.array([0.229,0.224,0.225])
    image = std * image + mean
    return np.clip(image, 0, 1)

# STEP 4: Upload your own images
from google.colab import files
print("Please upload your CONTENT image (your photo)")
content_upload = files.upload()
print("Please upload your STYLE image (painting)")
style_upload = files.upload()

content_path = list(content_upload.keys())[0]
style_path   = list(style_upload.keys())[0]

content = load_image(content_path, size=512)
style   = load_image(style_path, size=content.shape[-1])

plt.subplot(1,2,1); plt.imshow(im_convert(content)); plt.title("Content"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(im_convert(style)); plt.title("Style"); plt.axis("off")
plt.show()

# STEP 5: Load VGG19 for features
vgg = models.vgg19(weights=models.VGG19_Weights.IMAGENET1K_V1).features.to(device).eval()

def gram_matrix(x):
    b, c, h, w = x.size()
    x = x.view(c, h*w)
    return torch.mm(x, x.t())

def get_features(img):
    layers = {'0':'conv1_1','5':'conv2_1','10':'conv3_1','19':'conv4_1','21':'conv4_2'}
    feats, x = {}, img
    for name, layer in vgg._modules.items():
        x = layer(x)
        if name in layers: feats[layers[name]]=x
    return feats

# STEP 6: Extract features
content_feats = get_features(content)
style_feats   = get_features(style)
style_grams   = {l: gram_matrix(style_feats[l]) for l in style_feats}

# STEP 7: Create target (start from content image)
target = content.clone().requires_grad_(True).to(device)
opt = optim.Adam([target], lr=0.003)

# STEP 8: Optimize
steps = 500
for step in range(1, steps+1):
    t_feats = get_features(target)
    c_loss = torch.mean((t_feats['conv4_2']-content_feats['conv4_2'])**2)
    s_loss = 0
    for l in style_grams:
        t_gram = gram_matrix(t_feats[l])
        s_loss += torch.mean((t_gram - style_grams[l])**2)
    loss = c_loss + 1e6*s_loss
    opt.zero_grad(); loss.backward(); opt.step()
    if step % 100 == 0:
        print("Step", step, "Loss:", loss.item())
        plt.imshow(im_convert(target)); plt.title(f"Step {step}"); plt.axis("off"); plt.show()

# STEP 9: Save final result
plt.imshow(im_convert(target)); plt.axis("off")
plt.savefig("stylized_result.png")
print(" Stylized image saved as stylized_result.png")

from google.colab import files
files.download("stylized_result.png")
```

```python
# Install dependencies
!pip install torch torchvision Pillow

import torch, torch.nn as nn, torch.optim as optim
from torchvision import transforms, models
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt

# ========== Upload Images ==========
print("Upload your CONTENT image (photo)")
content_upload = files.upload()
print("Upload your STYLE image (painting)")
style_upload = files.upload()

content_img = list(content_upload.keys())[0]
style_img = list(style_upload.keys())[0]

# ========== Image Loader ==========
loader = transforms.Compose([
    transforms.Resize((256, 256)),  # keep it smaller for speed
    transforms.ToTensor()
])

def image_loader(image_name):
    image = Image.open(image_name)
    image = loader(image).unsqueeze(0)
    return image

content = image_loader(content_img)
style = image_loader(style_img)

# Show helper
def imshow(tensor, title=None):
    image = tensor.clone().detach().squeeze(0)
    image = transforms.ToPILImage()(image)
    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()

imshow(content, title="Content Image")
imshow(style, title="Style Image")

# ========== Model ==========
cnn = models.vgg19(weights=models.VGG19_Weights.DEFAULT).features.eval()

# Gram matrix
def gram_matrix(input):
    b, c, h, w = input.size()
    features = input.view(b * c, h * w)
    G = torch.mm(features, features.t())
    return G.div(b * c * h * w)

# Loss layers
class ContentLoss(nn.Module):
    def __init__(self, target):
        super().__init__()
        self.target = target.detach()
    def forward(self, x):
        self.loss = nn.functional.mse_loss(x, self.target)
        return x

class StyleLoss(nn.Module):
    def __init__(self, target_feature):
        super().__init__()
        self.target = gram_matrix(target_feature).detach()
    def forward(self, x):
        G = gram_matrix(x)
        self.loss = nn.functional.mse_loss(G, self.target)
        return x

# Normalization
normalization = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])

# Choose layers
content_layers = ['conv_4']
style_layers = ['conv_1','conv_2','conv_3','conv_4','conv_5']

# Build model
content_losses, style_losses = [], []
model = nn.Sequential(normalization)

i = 0
for layer in cnn.children():
    if isinstance(layer, nn.Conv2d):
        i += 1
        name = f'conv_{i}'
    elif isinstance(layer, nn.ReLU):
        name = f'relu_{i}'
        layer = nn.ReLU(inplace=False)
    elif isinstance(layer, nn.MaxPool2d):
        name = f'pool_{i}'
    else:
        continue
    model.add_module(name, layer)

    if name in content_layers:
        target = model(content).detach()
        cl = ContentLoss(target)
        model.add_module(f"content_loss_{i}", cl)
        content_losses.append(cl)
    if name in style_layers:
        target = model(style).detach()
        sl = StyleLoss(target)
        model.add_module(f"style_loss_{i}", sl)
        style_losses.append(sl)

for j in range(len(model)-1, -1, -1):
    if isinstance(model[j], (ContentLoss, StyleLoss)):
        break
model = model[:j+1]

# ========== Style Transfer ==========
input_img = content.clone()
optimizer = optim.LBFGS([input_img.requires_grad_()])
style_weight, content_weight = 1e6, 1
steps = 100

print("Optimizing...")
run = [0]
while run[0] <= steps:
    def closure():
        optimizer.zero_grad()
        model(input_img)
        style_score, content_score = 0, 0
        for sl in style_losses: style_score += sl.loss
        for cl in content_losses: content_score += cl.loss
        loss = style_weight*style_score + content_weight*content_score
        loss.backward()
        run[0] += 1
        if run[0] % 20 == 0:
            print(f"Step {run[0]} Style {style_score:.4f} Content {content_score:.4f}")
            imshow(input_img, title=f"Output at step {run[0]}")
        return loss
    optimizer.step(closure)

# Final output
print("Final Stylized Image:")
imshow(input_img, "Stylized Result")

# Save & download
out_img = transforms.ToPILImage()(input_img.squeeze(0).detach().cpu())
out_img.save("stylized_result.png")
files.download("stylized_result.png")
```
